# Combine the Clean Data

```python
# import libraries 
import requests
import pandas as pd
import numpy as np
import re

pd.set_option('display.max_columns', 500)
```

```python
# import the cleaned data 
main_df = pd.read_csv('../datasets/clean_data/florida_clean.csv')
ind = pd.read_csv('../datasets/clean_data/industry_clean.csv')
power = pd.read_csv('../datasets/clean_data/power_plants_clean.csv')
program = pd.read_csv('../datasets/clean_data/program_clean.csv')
irma = pd.read_csv('../datasets/clean_data/irma.csv')
ml = pd.read_csv('../datasets/clean_data/ml.csv')
age = pd.read_csv('../datasets/clean_data/age_clean.csv')
hospital = pd.read_csv('../datasets/clean_data/hospital_clean.csv')
```

```python
# check for missing values
print(main_df.isnull().sum().sum())
print(ind.isnull().sum().sum())
print(power.isnull().sum().sum())
print(program.isnull().sum().sum())
print(irma.isnull().sum().sum())
print(ml.isnull().sum().sum())
print(age.isnull().sum().sum())
print(hospital.isnull().sum().sum())
```

```python
# combine the data
df1 = pd.merge(main_df,ind, on='county')
df2 = pd.merge(df1,power, on='county', how='left')
df3 = pd.merge(df2,program, on='county', how='left')
df4 = pd.merge(df3, irma, on='county')
df5 = pd.merge(df4, ml, on='county', how = 'left')
df6 = pd.merge(df5, age, on='county')
final = pd.merge(df6, hospital, on='county')
```

```python
# replace missing values to zero and check the dataframe
final.fillna(0, inplace=True)
final
```

```python
# check to make sure the dataframe have all 67 counties.
len(final)
```

```python
# export the combined data
final.to_csv('../datasets/combined_final.csv', index = False)
```
